# Data Cleaning, Loading & Merging

```python
import pandas as pd
data = pd.read_csv('video_game.csv')  
```

```python
data.head(5)
```

```python
data = data.drop(['Unnamed: 0'], 1)
```

```python
data = data.set_index('Name')
```

## Reading in VG Sales Csv with Critic and User information

```python
vg_sales = pd.read_csv('vgsales.csv')  
```

```python
vg_sales = vg_sales.drop(['Rank'], 1)
```

```python
vg_sales = vg_sales.set_index('Name')
```

```python
data = data.merge(vg_sales, how='left', on='Name')
```

```python
data.describe()
```

```python
data.shape
```

## Reading in  All_Games.csv

```python
all_games = pd.read_csv('all_games.csv')  
```

```python
all_games.head(5)
```

```python
all_games = all_games.rename(columns={"name": "Name"})
```

```python
all_games.head(5)
```

```python
all_games = all_games.set_index('Name')
```

## Merging All_Games.csv with Game Names, Meta Score, and User Reviews

```python
data = data.merge(all_games, how='left', on='Name')
```

```python
data = data.drop(['Platform_y', 'Year','Genre_y','Publisher_y','platform','release_date','summary'], axis=1)
```

```python
data = data.rename(columns={"Platform_x": "Platform", "Genre_x": "Genre", "Publisher_x":'Publisher', "NA_players":'NA_Players','EU_players':'EU_Players','JP_players':'JP_Players','Other_players':'Other_Players','Global_players':'Global_Players','meta_score':'MetaCritic_Score','user_review':'MetaCritic_User_Score'})
```

## Cleaning Dataset

```python
data['Critic_Score'] = data['Critic_Score'].fillna(data['MetaCritic_Score'])
```

```python
data['User_Score'] = data['User_Score'].fillna(data['MetaCritic_User_Score'])
```

```python
data = data.drop(['MetaCritic_User_Score'], 1)
```

```python
data = data.drop(['Developer'], 1)
```

```python
data = data.drop(['Rating'], 1)
```

```python
data.isnull().sum()
```

```python
data = data.drop(['Critic_Count'], 1)
```

```python
data = data.drop(['User_Count'], 1)
```

```python
data = data.drop(['MetaCritic_Score'], 1)
```

```python
data.dtypes
```

```python
import numpy as np
data = data.replace('tbd', np.nan, regex=True)
```

```python
data['User_Score'] = data['User_Score'].astype(float)
```

```python
data.dtypes
```

```python
data.isnull().sum()
```

## Dropping NaN rows

```python
dropped_data_duplicates = data.dropna()
```

```python
dropped_data_duplicates = dropped_data_duplicates.drop_duplicates()
```

```python
dropped_data_duplicates = dropped_data_duplicates.reset_index()
```

## Dropping Duplicate Game Names

```python
dropped_name = dropped_data_duplicates.drop_duplicates(subset=['Name','Platform'], keep="first")
```

```python
dropped_name.isnull().sum()
```

```python
dropped_name['User_Score'] = dropped_data_duplicates['User_Score'] * 10
```

```python
dropped_name['Year_of_Release'] = dropped_data_duplicates['Year_of_Release'].astype(int)
```

```python
dropped_name = dropped_name.drop(['NA_Players', 'EU_Players','JP_Players','Other_Players','Global_Players'], axis=1)
```

## Feature Generation of Difference in Scores between Critics and Users

```python
dropped_name['Difference_In_Scores'] = dropped_name['User_Score'] - dropped_name['Critic_Score']
```

```python
def score_group(score):
    '''
    Function takes in critic/user score and divides them into groups of 10
    '''
    
    if score >= 90:
        return '90-100'
    elif score >= 80:
        return '80-89'
    elif score >= 70:
        return '70-79'
    elif score >= 60:
        return '60-69'
    elif score >= 50:
        return '50-59'
    else:
        return '0-49'
```

```python
dropped_name = dropped_name.dropna(subset=['Critic_Score']).reset_index()
dropped_name['Critic_Score_Group'] = dropped_name['Critic_Score'].apply(lambda x: score_group(x))
```

```python
def in_top(x):
    if x in pack:
        return x
    else:
        pass
def width(x):
    if x == 'Platform':
        return 14.4
    elif x == 'Developer':
        return 13.2
    elif x == 'Publisher':
        return 11.3
    elif x == 'Genre':
        return 13.6

def height(x):
    if x == 'Genre':
        return 8
    else:
        return 9
```

```python
cols = ['Genre', 'Publisher', 'Platform']
for col in cols:
    pack = []
    top = dropped_name[['Name', col]].groupby([col]).count().sort_values('Name', ascending=False).reset_index()[:15]
    for x in top[col]:
        pack.append(x)
    dropped_name[col] = dropped_name[col].apply(lambda x: in_top(x))
    dropped_data_duplicates_platform = dropped_name[[col, 'Critic_Score_Group', 'Global_Sales']].groupby([col, 'Critic_Score_Group']).median().reset_index().pivot(col, "Critic_Score_Group", "Global_Sales")
    plt.figure(figsize=(width(col), height(col)))
    sns.heatmap(dropped_data_duplicates_platform, annot=True, fmt=".2g", linewidths=.5).set_title((' \n'+col+' vs. critic score (by median sales) \n'), fontsize=18)
    plt.ylabel('', fontsize=14)
    plt.xlabel('Score group \n', fontsize=12)
    pack = []
```

```python
#dropped_data_duplicates = dropped_data_duplicates.dropna(subset=['Critic_Score']).reset_index()
dropped_name['User_Score_Group'] = dropped_name['User_Score'].apply(lambda x: score_group(x))
```

```python
cols = ['Genre', 'Publisher', 'Platform']
for col in cols:
    pack = []
    top = dropped_name[['Name', col]].groupby([col]).count().sort_values('Name', ascending=False).reset_index()[:15]
    for x in top[col]:
        pack.append(x)
    dropped_name[col] = dropped_name[col].apply(lambda x: in_top(x))
    dropped_data_duplicates_platform = dropped_name[[col, 'User_Score_Group', 'Global_Sales']].groupby([col, 'User_Score_Group']).median().reset_index().pivot(col, "User_Score_Group", "Global_Sales")
    plt.figure(figsize=(width(col), height(col)))
    sns.heatmap(dropped_data_duplicates_platform, annot=True, fmt=".2g", linewidths=.5).set_title((' \n'+col+' vs. user score (by median sales) \n'), fontsize=18)
    plt.ylabel('', fontsize=14)
    plt.xlabel('Score group \n', fontsize=12)
    pack = []
```
